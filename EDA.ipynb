```python
# Calculate the number of matching values between the two columns
matches = (data['favorite_category_meta_category'] == data['favorite_item_meta_category']).sum()

# Calculate the percentage of matching values
total_rows = len(data)
match_percentage = (matches / total_rows) * 100

print(f"Number of matching values: {matches}")
print(f"Percentage of matching values: {match_percentage:.2f}%")
```

```python
# Create the 'avg_tips' column
data['avg_tips'] = data['total_tips'] / data['total_orders']

# Display the 50 largest avg_tips
largest_avg_tips = data.nlargest(50, 'avg_tips')
print("Top 50 largest average tips:")
print(largest_avg_tips)
```

```python

# Display the 50 smallest avg_tips
smallest_avg_tips = data.nsmallest(50, 'avg_tips')
print("\nTop 50 smallest average tips:")
print(smallest_avg_tips)
```

```python
# Assuming 'data' is your DataFrame with the 'total_orders' column

plt.figure(figsize=(10, 6))
plt.hist(data['total_orders'], bins=100, edgecolor='black')
plt.xlabel('Total Orders')
plt.ylabel('Frequency')
plt.title('Histogram of Total Orders')
plt.show()
```

```python
plt.figure(figsize=(8, 6))
plt.boxplot(data['total_orders'], vert=False)
plt.xlabel('Total Orders')
plt.title('Boxplot of Total Orders')
plt.show()
```
